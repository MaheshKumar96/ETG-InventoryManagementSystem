# Reading Record File

```python
import json

fd = open("record.json",'r')
data = fd.read()
fd.close()

products = json.loads(data)
```

# Adding new products to Inventory

```python
print("===============================")
print("       Welcome To MK Store        ")
print("===============================")
print("Add Product To Inventory - ")
print("")
product_id = int(input("Enter Product ID: "))
name = str(input("Enter Product Name : "))
quantity = int(input("Enter Quantity : "))
price = int(input("Enter Price : "))
category = str(input("Enter Product Category : "))
discount = int(input("Enter Discount : "))
print("")
print("===============================")
print("Product Added Successfully ")
print("===============================")

products[product_id] = {"Name": name, "Quantity": quantity, "Price": price, "Category": category, "Discount": discount}

js = json.dumps(products)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
